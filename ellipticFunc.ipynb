```python
import numpy as np
import matplotlib.pyplot as plt
hani = 1
k=0.7
points = 1000
```

```python
x = np.arange(0,hani,hani/points)
su = np.zeros(points)
for i in range(len(x)):
  for j in range(i):
    su[i] += 1/( points/hani* np.sqrt((1-x[j]**2)*(1-(k*x[j])**2)))

new = np.linspace(0,su[-1],points)
sn = np.interp(new, su, x)
cn = np.sqrt(1-sn**2)
dn = np.sqrt(1-(k*sn)**2)
plt.plot(x,sn)
plt.plot(x,cn)
plt.plot(x,dn)
plt.show()
```
